This Note book extracts papers from PubMed and BioMed and put the results in a dataframe. Helpful functions for expracting papers from PubMed is taken from https://gist.github.com/bonzanini/5a4c39e4c02502a8451d 

```python
from Bio import Entrez
from Bio import Medline
import json
import pandas as pd
import requests
```

```python
def esearch(query):
    Entrez.email = 'bkjboppana@gmail.com'
    search = Entrez.esearch(db='pubmed', 
                            sort='relevance', 
                            retmax='5000',
                            retmode='xml', 
                            term=query)
    papers = Entrez.read(search)
    return papers 

def efetch(idlist):
    ids = ','.join(idlist)
    Entrez.email = 'bkjboppana@gmail.com'
    fetch = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids)
    papers = Entrez.read(fetch)
    return papers
```

```python
results = esearch('Parkinson')  #searching ID using esearch from PubMed API
id_list = results['IdList']     
papers = efetch(id_list)  #Fetching papers in XML format using efetch function
```

```python
#Creating empty lists to insert paper date,keywords, abstract and title
date =[]
keys=[]
abstract=[]
title=[]
for i,paper in enumerate(papers['PubmedArticle']):

    try:
            date.append(json.dumps(paper['MedlineCitation']["DateRevised"]))
            keys.append(paper['MedlineCitation']["KeywordList"])
            title.append(paper['MedlineCitation']['Article']['ArticleTitle'])
            abstract.append(paper['MedlineCitation']['Article']['Abstract']['AbstractText'])
    except KeyError:
            pass
       
#json.dumps(papers['PubmedArticle'][1]['MedlineCitation'])
#json.dumps(papers['PubmedArticle'][1]['MedlineCitation']['Article']['ArticleTitle'])
```

```python
#Inserting columns into Dataframe
pubmed_df = pd.DataFrame(columns = ['Date','Keywords','Title','Abstract'])
pubmed_df['Date'] = pd.Series(date)
pubmed_df['Keywords'] = pd.Series(keys)
pubmed_df['Title'] = pd.Series(title)
pubmed_df['Abstract'] = pd.Series(abstract)
```
